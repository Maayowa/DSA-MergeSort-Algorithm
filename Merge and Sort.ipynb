## Merge Sort (Divide and conquer Algorithm)

```python
def mergesort(nums):
    """
    This is the part where the division of the list is done
    """
    # Condition for ending recursion
    if len(nums) < 2:
        return nums
    
    mid = len(nums) // 2
    
    # Splitting list into halves
    
    left = nums[:mid]
    right = nums[mid:]
    
    
    # solving and breaking down lists by recursion
    
    left_sort, right_sort = mergesort(left), mergesort(right)
    
    sorted_nums = merge(left_sort, right_sort)
    
    
    return sorted_nums


def merge(nums1, nums2):
    """
    This is the 'conquer function where smaller lists are merged'
    """
    # Allocating pointers for looping through and sorting the list
    nest_list = []
    i = 0
    j = 0
    
    while i < len(nums1) and j < len(nums2):
        if nums1[i] < nums2[j]:
            nest_list.append(nums1[i])
            i += 1
        else:
            nest_list.append(nums2[j])
            j += 1
            
    # After looping and sorting
    nums1_rem = nums1[i:]
    nums2_rem = nums2[j:]
    
    return nest_list + nums1_rem + nums2_rem
    
```

```python
merge([1,4,7,9,11], [-1, 0, 2, 3, 6, 12])
```
